## Clean Version, langsung kesini

```python
v = clean.username.value_counts()
filter=clean[clean.username.isin(v.index[v.gt(0)])] #1 kalau threshold nya 1, ganti sesuai keinginan
cleanth=pd.DataFrame(filter)
cleanth.info()
```

```python
import networkx as nx
clean_graph = nx.from_pandas_edgelist(cleanth,source="username",target="mentions")
```

```python
import matplotlib.pyplot as plt
leaderboard = {}
for x in clean_graph.nodes:
    leaderboard[x] = len(clean_graph[x])
s = pd.Series(leaderboard, name='connections')
connect = s.to_frame().sort_values('connections', ascending=False)
connect.head(15)
#connect.to_csv('con_event9_speaker2')
```

```python
connect=connect.reset_index()
labelnode=connect['index'].iloc[pd.np.r_[0:]]# ini isi kalo mau skip nama2 di atas, kaya tedx kalo mau di skip isi di iloc nya dari perapa sampe berapa, multiple juga bisa
listlabel= labelnode.values.tolist() ## lanjutan diatas, 1:91 kalau jumlahnya ada 91, lihat di cleanth.info harus didefine
listlabel #lihat list dibawah untuk melihan nama yg mau di skip
```

```python
labels = {}
for character in listlabel:
    labels[character] = character
```

```python
colors = [i/len(clean_graph.nodes) for i in range(len(clean_graph.nodes))]
```

```python
edgecol= [i/len(clean_graph.edges) for i in range(len(clean_graph.edges))]
```

```python
plt.figure(figsize=(100,100))
d = dict(clean_graph.degree)
pos_tweet=nx.spring_layout(clean_graph)

nx.draw_networkx_nodes(clean_graph, with_labels=False,node_color=colors,cmap=plt.cm.viridis_r, node_size=[v * 3000 for v in d.values()],pos=pos_tweet, font_color='r')
nx.draw_networkx_edges(clean_graph, with_labels=False,edge_color=edgecol,connectionstyle='arc3, rad=0.2',pos=pos_tweet, width=8, edge_cmap=plt.cm.spring , font_color='r',arrows=True,arrowstyle='-|>', arrowsize=20)
nx.draw_networkx_labels(clean_graph, with_labels=True,labels=labels,font_size=25,pos=pos_tweet, font_color='#DC143C', font_weight='bold')
plt.savefig("event11_speaker6.png", dpi = 250,bbox_inches='tight', transparent=True)
plt.axis('off')
plt.show()
```
